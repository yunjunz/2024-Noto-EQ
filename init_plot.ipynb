## Initial plot of the dense offset results from `alos2App`

```python
%matplotlib inline
import os
import glob
import datetime as dt
import numpy as np
import platform
from cartopy import crs as ccrs
from matplotlib import pyplot as plt
from mintpy.utils import ptime, readfile, writefile, isce_utils, utils as ut, plot as pp
from mintpy.cli import geocode, load_data, multilook, smallbaselineApp, view
plt.rcParams.update({'font.size': 12})

# platform
if platform.system() == 'Linux':
    proj_dir = os.path.expanduser('~/data/2024NotoPenEQ')
else:
    proj_dir = os.path.expanduser('~/data/archives/2024NotoPenEQ')

# work directory
work_dir = os.path.join(proj_dir)
os.chdir(work_dir)
print('Go to directory:', work_dir)
```

```python
def get_file_names(proj_dir, ds_name):
    date1, date2 = ptime.yymmdd(ds_name.split('_')[-2:])
    os.makedirs(os.path.join(proj_dir, ds_name, 'pic'), exist_ok=True)
    # input / output
    off_file = os.path.join(proj_dir, ds_name, f'dense_offset/{date1}-{date2}_denseoffset.off')
    png_file = os.path.join(proj_dir, ds_name, f'pic/dense_offset.png')
    return off_file, png_file
```

```python
off_file, png_file = get_file_names(proj_dir, ds_name='ALOS2_A121_20220926_20240101')
view.main(f'{off_file} -v -2 2 --flip-lr --flip-ud --sub-y 550 1050 --zero-mask --noverbose -o {png_file}'.split())
```
